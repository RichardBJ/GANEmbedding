```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
```

```python
tsne_df = pd.read_csv("TSNE.csv", index_col = 0)
tsne_df.head()
```

```python
umap_df = pd.read_csv("umap.csv", index_col = 0)
umap_df.head()
```

```python
cmap = get_cmap("tab20")

print(cmap(0))
print(cmap(0.05))
print(cmap(0.1))
```

```python
# Let's get the data into a more clean form.

class SingleChannelDataWrapper:
    
    def __init__(self, name, gan_x, gan_y, raw_x, raw_y):
        
        self.name = name
        self.gan_x = gan_x
        self.gan_y = gan_y
        self.raw_x = raw_x
        self.raw_y = raw_y
        

def build_aggregate_viz(dataset, model_name):
    
    # Build the large visualisation with all points
    
    # First step - put all the data in the wrapper above to make life easy.
    data = get_all_data(dataset)
    
    fig, ax = plt.subplots(figsize=(10, 10))
    
    
    for idx, channel in enumerate(data):
        
        # Hacky colour solution
        cols = [cmap(idx/10) for _ in channel.gan_x]
        cols_two = [cmap(idx/10 + 0.05) for _ in channel.gan_x]
        
        ax.scatter(channel.gan_x, channel.gan_y, c=cols, cmap="tab20b", s=7, alpha=0.8, label=f"{channel.name} GAN Data")
        ax.scatter(channel.raw_x, channel.raw_y, c=cols_two, cmap="tab20b", s=7, alpha=0.8, label=f"{channel.name} Raw Data")
    
    ax.legend()
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    
    ax.set_title(f"{model_name} comparison graph of GAN generated and corresponding raw data.")
    plt.show()
    
def build_comparison_viz(dataset, model_name):
    
    data = get_all_data(dataset)
    
    fig, ax = plt.subplots(5, 1, figsize=(10, 50))
        
    for idx, channel in enumerate(data):
        
        # Hacky colour solution again
        cols = col_generator(idx, channel)
        
        ax[idx].scatter(channel.gan_x, channel.gan_y, c=cols[0], s=7, alpha=0.8, label=f"GAN Data")
        ax[idx].scatter(channel.raw_x, channel.raw_y, c=cols[1], s=7, alpha=0.8, label=f"Raw Data")
        
        ax[idx].legend()
        ax[idx].set_title(f"Isolated pairwise {model_name} comparison of GAN and raw data for {channel.name} channel")
        
    plt.show()
    

def get_all_data(dataset):
    
    pyr = get_data(dataset, "4096PYR")
    edr = get_data(dataset, "4096edr01")
    lina = get_data(dataset, "4096lina11")
    cnf = get_data(dataset, "c035[NF=50Hz]0-32_4k")
    pvn = get_data(dataset, "pvn")
    
    return (pyr, edr, lina, cnf, pvn)
    
    

def get_data(dataset, prefix):
    
    return SingleChannelDataWrapper(prefix, dataset[f"{prefix}GAN_X"], dataset[f"{prefix}GAN_Y"], dataset[f"{prefix}raw_X"], dataset[f"{prefix}raw_Y"])
        

def col_generator(idx, channel):
    
    cmap = get_cmap("tab20")

    return ([cmap(idx/10) for _ in channel.gan_x], [cmap(idx/10 + 0.05) for _ in channel.gan_x])
```

```python
build_aggregate_viz(tsne_df, "T-SNE")
```

```python
build_aggregate_viz(umap_df, "UMAP")
```

```python
build_comparison_viz(tsne_df, "T-SNE")
```

```python
build_comparison_viz(umap_df, "UMAP")
```
